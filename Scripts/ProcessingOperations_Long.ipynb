```python
# Procedures to execute
DataPreparation = 'yes'
DataStage1 = 'yes'
DataStage2_EMD = 'yes'
DataStage3_PGA = 'yes'
```

```python
# Raw data directory (.fif file).
SourceDataDir = "D:\\Data\OCT_TDT\pre_stimuli\Stage4\data"

# File names to prepare for analysis.
DataFileNames = "all"
```

```python
##############################|-------> Settings <-------|##############################
Settings = {}
# |------- Channels -------| #
Settings["badChannels"] = [18, 21]              # Default: [18, 21]
Settings["numberOfChannels"] = 64               # Default: 64
Settings["eventChannels"] = range(64,68)        # Default: range(64,68)

# |------- Storage -------| #
Settings["fullDump"] = 'no'                     # Default: 'no'

# |------- Filtering -------| #
Settings["RawDataFiltration"] = 'yes'
Settings["cutoff_l"] = 0.1                      # Default: 0.1
Settings["cutoff_h"] = 40                       # Default: 40

Settings["DT"] = 'auto'                         # Default: 'auto'
Settings["DT_param"] = None                     # Default: None

# |------- Epoching -------| #
Settings["epochTypes"] = ["010", "011", "100", "101", "110", "111"]

Settings["resemple"] = 'yes'                    # Default: 'yes'
Settings["new_freq"] = 500                      # Default: 500

Settings["tmin"] = 5.8                       # Default: -0.2
Settings["tmax"] = 6.0                         # Default: 0.8
Settings["baseline"] = (5.8, 5.8)                # Default: (-0.2, 0)

Settings["min_event_duration"] = 0.1            # Default: 0.1
Settings["min_voice_duration"] = 0.01           # Default: 0.01

# |------- Epoch Outliers -------| #
Settings["dropOutliers"] = 'yes'                # Default: 'yes'
Settings["stdBand"] = 2                         # Default: 2
# Outlier detection band based on: mu +/- stdBand * sigma
# stdBand coverage in percents: 1 - 68.2%, 2 - 95.4%, 3 - 99.7%
Settings["scoreTh"] = 75                        # Default: 75
# Each epoch is given a score based on number of outlier samples. Any epoch with a score lower than scoreTh is marked as an outlier epoch.

# |------- Initialization -------| #
Settings["DeviceNormalization"] = 'yes'         # Default: 'yes'
Settings["rangeFactor"] = 100/4.5               # Default: 100/4.5
Settings["coilType"] = 5001                     # Default: 5001

Settings["grad"] = 7000e-13                     # Default: 7000e-13
Settings["preload"] = True                      # Default: True

Settings["rectifyEvents"] = 'yes'               # Default: 'yes'
Settings["eventAmplitudeTh"] = 1                # Default: 1

Settings["fixEventSequence"] = 'yes'            # Default: 'yes'
Settings["exposureTime"] = 300                  # Default: 300
Settings["exposure_err"] = 0.15                 # Default: 0.15

Settings["reconfigureEvents"] = 'yes'           # Default: 'yes'
Settings["useRecreatedEvents"] = 'yes'          # Default: 'yes'

Settings["delayTime"] = 8000                    # Default: 8000
Settings["errorMargin"] = 0.01                  # Default: 0.01

Settings["dropFaultyResponseEpochs"] = 'no'     # Default: 'no'
Settings["voiceOutliers"] = 'no'                # Default: 'no'

# Naming pattern
Settings["idStartInd"] = 3                      # Default: 3
Settings["idEndInd"] = 7                        # Default: 7
Settings["dateStartInd"] = 8                    # Default: 8
Settings["dateEndInd"] = 14                     # Default: 14

# |------- EMD -------| #
Settings["cycles"] = 5                          # Default: 5

# |------- PGA pre-load settings -------| #
Settings["loadEMD"] = 'yes'                     # Default: 'yes'
# Use 'yes' when loading Stage2_EMD file and 'no' when loading Stage1 file.

Settings["getFilt"] = 'emd_filter1'             # Default: 'emd_filter1'
# getFilt options are:
# 'base' - Returns the preprocessed data
# 'emd_filter1' - Returns data filtered using EMD. Reduction of high frequency noise.
# 'emd_filter2' - Returns data filtered using EMD. Reconstruction of low frequency IMFs.
# 'IMF#' - Replace # with a number (e.g. 'IMF3'). Returns specific IMF's data

Settings["getFullDump"] = 'no'                  # Default: 'no'
# Only use 'yes' if  the dump setting was Settings["fullDump"] = 'yes'. 'no' may be used regardless of that setting. Is set to 'yes' the Load_EMD_Dataset function will return 3 parameters: epochs_manager, events_conditions and voiceResponsesInfo.

# |------- PGA settings -------| #
Settings["numberOfElements"] = 5                # Default: 5
# numberOfElements defines how many events are averaged together
Settings["maxIntersection"] = 2                 # Default: 2
# maxIntersection defines the maximum intersection of events between different average groups
Settings["numberOfAttempts"] = 10               # Default: 10
# numberOfAttempts defines number of attempts to find the largest combination set
Settings["TDT"] = "clustered1"                     # Default: 'pga_ref'
# TDT can be set to "classic", "pga_gauss", pga_ref or "known". "classic" will not perform PGA on the Test dataset. "pga_gauss" and "pga_ref" will perform separate PGA procedure for train and test sets using different reference sources. "known" will perform a single PGA procedure for the entire data (cannot surve as proof for classification).
Settings["TrSize"] = 'max'                     # Default: 'max'
# TrSize can be set to value between 0 and 100 for Settings["TDT"] = "classic". If set to numeric value, for example Settings["TrSize"] = 80, train set will be 80% and test set will be 20% of the data.
# TrSize can be set to 'same' or 'max' for other "TDT" settings. If set to 'same', test and train sets will have the same size. If set to 'max', train might be larger than the test.
```

```python
# Load project library.
ex = WU_MEG_DP.Sumitomo_Decoding_Long(Settings)
Settings["Sumitomo_Long_Short"] = "long"
```

```python
# Directory selection.
Settings["inDataDir"] = SourceDataDir
Settings["outDataDir"] = SourceDataDir + "_Raw"
Settings["DataFileNames"] = DataFileNames

# Preparation of the datafile "00". Not required if was calculated before.
if DataPreparation == 'yes':
    ex.PrepareRawData(Settings["inDataDir"], Settings["DataFileNames"], Settings["outDataDir"])
```

```python
# Directory selection.
Settings["inDataDir"] = SourceDataDir + "_Raw"
Settings["outDataDir"] = SourceDataDir + "_Stage1"
if DataFileNames != "all":
    for k in range(len(DataFileNames)):
        DataFileNames[k] = DataFileNames[k][:Settings["idEndInd"]] + "_00_raw.fif"
    Settings["DataFileNames"] = list(set(DataFileNames))
else:
    Settings["DataFileNames"] = DataFileNames

# Preparation of stage1- preprocessed data.
if DataStage1 == 'yes':
    ex.PrepareStage1Data(Settings)
```

```python
# Directory selection.
Settings["inDataDir"] = SourceDataDir + "_Stage1"
Settings["outDataDir"] = SourceDataDir + "_Stage2_EMD"
if DataFileNames != "all":
    for k in range(len(DataFileNames)):
        DataFileNames[k] = DataFileNames[k][:Settings["idEndInd"]] + "_Stage1.pkl"
    Settings["DataFileNames"] = list(set(DataFileNames))
else:
    Settings["DataFileNames"] = DataFileNames

# Preparation of stage2- EMD data.
if DataStage2_EMD == 'yes':
    ex.PrepareStage2EMDdata(Settings)
```

```python
# Directory selection.
Settings["inDataDir"] = SourceDataDir + "_Stage2_EMD"
Settings["outDataDir"] = SourceDataDir + "_PGA"
if DataFileNames != "all":
    for k in range(len(DataFileNames)):
        DataFileNames[k] = DataFileNames[k][:Settings["idEndInd"]] + "_Stage2_EMD.pkl"
    Settings["DataFileNames"] = list(set(DataFileNames))
else:
    Settings["DataFileNames"] = DataFileNames

if DataStage3_PGA == 'yes':
    ex.PrepareDataForML(Settings)
```
